```python
import requests
import json
from time import sleep
import pandas as pd

# result = []
n = 1
for i in range(15):
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36 Edg/109.0.1518.61'
    }
    params = {
        "batch":1,
        "input":'{"0":{"json":{"skip":' + str(i*36) + ',"tag":null,"sort":null,"q":null,"language":"en"},"meta":{"values":{"tag":["undefined"],"sort":["undefined"],"q":["undefined"]}}}}'
    }
    proxies = {
        'http': 'http://127.0.0.1:10810',
        'https': 'http://127.0.0.1:10810'
    }

    url = "https://flowgpt.com/api/trpc/prompt.getPrompts"
    response = requests.get(url=url,params=params,headers=headers,verify=False,proxies=proxies)
    res = response.json()[0]['result']['data']['json']

    for j in range(len(res)):
        # result.append([res[j]['title'],res[j]['uses'],res[j]['comments'],res[j]['initPrompt']])
        with open("ChatGPT_Prompt.txt",'a',encoding='utf-8') as f:
            f.write("title：" + res[j]['title'] + "\n")
            f.write("uses：" + str(res[j]['uses']) + "\n")
            f.write("comments：" + str(res[j]['comments']) + "\n")
            f.write("initPrompt：\n" + res[j]['initPrompt'] + "\n-------------------------------以上为第{}条---------------------------------\n".format(n))
            n += 1
            
# df = pd.DataFrame(result,columns=['title','uses','comments','initPrompt'])
# df.to_csv('prompt.csv')
```
